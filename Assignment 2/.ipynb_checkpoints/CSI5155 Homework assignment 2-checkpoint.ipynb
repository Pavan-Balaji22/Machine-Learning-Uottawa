```python
import pandas as pd
import numpy as np
```

```python
# Importing bank dataset
bank_dataset = pd.read_csv("bank-additional/bank-additional-full.csv",sep=";")
bank_X = bank_dataset.iloc[:,:-1] # Features
bank_y = bank_dataset.iloc[:,-1] # Target variable
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
def cat_to_num(X, to_one_hot = [], to_ordinal =[]):
    '''
    Input:  X - original dataset
            to_one_hot - the columns that need to be One hot encoded
            to_ordinal  - the columns that need to be oridinal encoded
    
    Output: X - after all the processing done
    
    Function to change the categorical variables present in the dataset to encoded numericals   
    
    '''
    OHen = OneHotEncoder()
    Oren = OrdinalEncoder()
    n = X.shape[1]
    # looping through the columns and encodering the categorical column     
    for i in range(n):
        
        # One hot encoding for cardinal columns
        if X.iloc[:,i].name in to_one_hot:
            OHE = OHen.fit_transform(X.iloc[:,[i]])
            df = pd.DataFrame(OHE.toarray(),columns=[X.iloc[:,i].name+"_"+y for y in X.iloc[:,i].unique()])
            X = pd.merge(left=X,right=df,left_index=True,right_index=True)
        
        # Ordinal encoding for ordinal columns
        if X.iloc[:,i].name in to_ordinal:
            X.loc[:,X.iloc[:,i].name] = Oren.fit_transform(X.iloc[:,[i]])
    
    # Dropping the cardinal categorical columns to avoind redundancy  
    X.drop(columns=to_one_hot,inplace = True)
    return X
```

```python
# Converting test data to numerical data

ordinal = ['day_of_week','month','contact']
cardinal =['job','marital','education','default','housing','loan','poutcome']
bank_X = cat_to_num(bank_X,to_one_hot=cardinal,to_ordinal=ordinal)
bank_X.head()
```

## Part A - Class imbalance, ensembles, and feature selection

### 1. Performing Oversampling, Under-sampling, Balanced sampling

```python
# Performing Oversampling on Bank data
from imblearn.over_sampling import SMOTE

over_sampling = SMOTE(sampling_strategy="minority",random_state=123)
over_sampled_X,over_sampled_y = over_sampling.fit_sample(bank_X,bank_y)
```

```python
print("Before Sampling:\n")
print("Shape of Original Dataset : {} {}".format(*bank_X.shape))
print("No of Instances in class {} : {}".format("Yes",len(bank_y[bank_y=="yes"])))
print("No of Instances in class {} : {}\n".format("No",len(bank_y[bank_y=="no"])))
print("After Over Sampling:\n")
print("Shape of Over sampled dataset : {} {}".format(*over_sampled_X.shape))
print("No of Instances in class {} : {}".format("Yes",len(over_sampled_y[over_sampled_y=="yes"])))
print("No of Instances in class {} : {}".format("No",len(over_sampled_y[over_sampled_y=="no"])))
```

```python
# Performing undersampling on Bank data
from imblearn.under_sampling import NearMiss

under_sampling = NearMiss(sampling_strategy="majority")
under_sampled_X,under_sampled_y = under_sampling.fit_sample(bank_X,bank_y)
```

```python
print("Before Sampling:\n")
print("Shape of Original Dataset : {} {}".format(*bank_X.shape))
print("No of Instances in class {} : {}".format("Yes",len(bank_y[bank_y=="yes"])))
print("No of Instances in class {} : {}\n".format("No",len(bank_y[bank_y=="no"])))
print("After Under Sampling:\n")
print("Shape of Under sampled dataset : {} {}".format(*under_sampled_X.shape))
print("No of Instances in class {} : {}".format("Yes",len(under_sampled_y[under_sampled_y=="yes"])))
print("No of Instances in class {} : {}".format("No",len(under_sampled_y[under_sampled_y=="no"])))
                                                                                              
```

```python
# Performing Balanced sampling on Bank data
from imblearn.combine import SMOTETomek

balanced_sampling = SMOTETomek(sampling_strategy="auto",random_state=123)
balanced_sampled_X,balanced_sampled_y = balanced_sampling.fit_sample(bank_X,bank_y)
```

```python
print("Before Sampling:\n")
print("Shape of Original Dataset : {} {}".format(*bank_X.shape))
print("No of Instances in class {} : {}".format("Yes",len(bank_y[bank_y=="yes"])))
print("No of Instances in class {} : {}\n".format("No",len(bank_y[bank_y=="no"])))
print("After Balanced Sampling:\n")
print("Shape of Balanced sampled dataset : {} {}".format(*balanced_sampled_X.shape))
print("No of Instances in class {} : {}".format("Yes",len(balanced_sampled_y[balanced_sampled_y=="yes"])))
print("No of Instances in class {} : {}".format("No",len(balanced_sampled_y[balanced_sampled_y=="no"])))
                                                                                              
```

### 6. Determining Satistical significant difference in the accuracies obtained by the six algorithm  


```python
from scipy.stats import ttest_rel

def calculate_p_values(data):
    '''
    Input:
        data: 
            Data obtained from crossvalidation fro the models in consideration
    Output:
        p_values: 
            The p_values obtained when performing paired t-test on the models in consideration
        differences: 
            The Dataframe show results of the paired t-test and comparisions
    
    Function to perform t_test and calculate p-value 
    '''
    # Declaring Necessary variables
    p_values = list()
    mem = list()
    differences = list()
    models = list()
    
    # Performing t-test on all the models
    for i in range(data.shape[1]-1):
        for j in range(data.shape[1]):
            if i!=j:
                if ((j,i) not in mem):
                    
                    # Performing t-test on odel a and b
                    a,b =data.iloc[:,i],data.iloc[:,j]
                    _,p_value = ttest_rel(a,b)
                    
                    differences.append(np.array(a-b)) # appending difference between accuracy of model a and b in the ten folds
                    p_values.append(p_value) # appending p_value from the t-test
                    mem.append((i,j)) # appending the combination compared avoid repetition
                    models.append(data.iloc[:,i].name+"-"+data.iloc[:,j].name) # appeding names for columns
    
    # Merges the Data collected and Statistics on the data
    differences = np.c_[differences].T
    differences = np.r_[differences,
                        np.mean(differences,axis=0).reshape((1,differences.shape[1])),
                        np.std(differences,axis=0).reshape((1,differences.shape[1])),
                        np.array(p_values).reshape((1,differences.shape[1]))]
    # Creating Dataframe
    index = [*range(1,11),"Average Accuracy(mean)","Standard Deviation","p-value"]
    differences = pd.DataFrame(differences,index=index,columns=models)
    differences.index.set_names("Fold",inplace=True)
    differences.columns.set_names("Accuracy Difference",inplace=True)
    
    
    return p_values,differences
```

```python
# Calculating p_values and creating the table required
p_values,data_frame= calculate_p_values(K_fold_table.iloc[:10,:])

# Printing the p_values calculated
print("The p_values calculated for the models :\n")
print("\n".join(["{} is {:.4f}".format(x,y) for x,y in zip(data_frame.columns.values,p_values)]))
```

```python
# Displaying the table
data_frame.style.set_properties(**{"text-align" : "center"})
data_frame.style.set_properties(pd.IndexSlice[10],**{"border-bottom" : "solid","border-width": "2px"})
```

```python
# the pairs of models with statistical significant difference
print("From the table it is clear that there is statistical significant difference between the following models:\n")

for x,y in zip(data_frame.columns.values,p_values):
    if y < 0.005:
        algo=x.split("-")
        print("{} and {}".format(algo[0],algo[1]))
```

### Calulating satistical significance using Friedman's Test

```python
def Friedman_test(Dataset,names):
    
    n,k=Dataset.shape
    
    Rank_table = Dataset.rank(axis=1,method="max",ascending=False)
    average_ranks=np.mean(Rank_table,axis=0)
    
    R_d = sum((avg_ranks*n))/(n*k)
    ssd1 = n*sum((avg_ranks - R_d)**2)
    ssd2 = sum(np.sum((Rank_table-R_d)**2).values)/(n*(k-1))
    
    F_stat = ssd1/ssd2 
    
    return F_stat,average_ranks
```

```python
Friedman_stat,avg_ranks = Friedman_test(Dataset=Results_table,names=Algorithms)
```

```python
print("The Friednman Statistic obtained: {}".format(Friedman_stat))
print("The Average ranks from Friedman's Test:\n")
print(avg_ranks)
```

From the results, It is clear that there is no Statistically significant difference between K Nearest neghbours,Random Forest and Support vector machines since the Friedman statictic 1.5 is less the the critical value 6.5 from the Friedman table for k = 3 and n =4.
